```python
X_train, X_test, y_train, y_test = train_test_split(diabetes['data'],
                                                        diabetes['target'],
                                                        random_state=0)
```

```python
# Transform into numpy arrays for easier manipulations:
X_train = np.array(X_train)
X_test = np.array(X_test)
y_train = np.array(y_train)
y_test = np.array(y_test)
```

```python
X_b = np.c_[np.ones((331, 1)), X_train] # add x0 = 1 to each instance, shape 331 x 11
theta_best = np.linalg.inv(X_b.T.dot(X_b)).dot(X_b.T).dot(y_train)  # normal equation
```

```python
print(theta_best)
```

```python
X_test_b = np.c_[np.ones((111, 1)), X_test]
```

```python
y_pred = X_test_b.dot(theta_best)
```

```python
# Measure the model:
mse = mean_squared_error(y_test, y_pred)
r = r2_score(y_test, y_pred)
```

```python
print(mse)
print(r)
```

```python
# compare with sklearn linear regression model:
lin_reg = LinearRegression()
reg = lin_reg.fit(X_train, y_train)
print("Intercept={}, Slope={}".format(lin_reg.intercept_, lin_reg.coef_))
```

Batch Gradient Descent:

```python
etta = 0.1 # learning rate
n_iterations = 100
m=len(y_train)

theta = np.random.randn(11,1) # random initialization <11x1>

for iteration in range(n_iterations):
    predictions = np.array(X_b.dot(theta))
    difference = np.subtract(predictions, y_train.reshape(-1,1))
    gradients = 2/m * X_b.T.dot(difference)
    theta = theta - etta * gradients
print(theta)
```

Comparing batch gradient descent and the previous results:

```python
y_pred_grad = X_test_b.dot(theta)
```

```python
mse = mean_squared_error(y_test, y_pred_grad)
r = r2_score(y_test, y_pred_grad)
print(mse, r)
```

It's quite a bit worse. Try to improve with grid search:

```python
def gradient_descent_prediction(learning_rate: float) -> float:
    etta = learning_rate
    n_iterations = 100
    m=len(y_train)

    theta = np.random.randn(11,1) # random initialization <11x1>
    y_pred = 0

    for iteration in range(n_iterations):
        predictions = np.array(X_b.dot(theta))
        difference = np.subtract(predictions, y_train.reshape(-1,1))
        gradients = 2/m * X_b.T.dot(difference)
        theta = theta - etta * gradients
        y_pred = X_test_b.dot(theta)
    return y_pred
```

```python
best_score = 0
best_parameter = 100000
# note the values have been tuned in the list starting from the
# general [0.001, 0.01, 0.1, 1, 10, 100].
for etta in [0.9, 0.92, 0.94, 0.96, 0.98, 1]:
    # for each combination of parameters, evaluate the model
    y_pred = gradient_descent_prediction(etta)
    new_score = r2_score(y_test,y_pred)
    if new_score > best_score:
        best_score = new_score
        best_parameter = etta
print("Best score: {:.7f}".format(best_score))
print("Best parameters: {}".format(best_parameter))
```

The results are now comparable to the other methods.
How close can it get with stochastic gradient descent?

```python
n_epochs = 15000
t0, t1 = 5, 50 # learning schedule hyper-parameters

def learning_schedule(t):  # keeping this since we have not seen others.
  return t0/(t+t1)

theta = np.random.randn(11,1) # random initialization

for epoch in range(n_epochs):
  for i in range(m):
    random_index = np.random.randint(m)
    xi = X_b[random_index:random_index+1]
    yi = y_train[random_index:random_index+1]
    gradients = 2 * xi.T.dot(xi.dot(theta) - yi)
    eta = learning_schedule(epoch * m + i)
    theta = theta - eta * gradients
y_pred_stochastic = X_test_b.dot(theta)
print(r2_score(y_test,y_pred_stochastic))
```

To obtain the same result with this setup we need to crank
up the epochs by quite a bit, but doing so we can get close.
