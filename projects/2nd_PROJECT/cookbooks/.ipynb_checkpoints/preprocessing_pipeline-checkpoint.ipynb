### Outliers

```python
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder, RobustScaler
from sklearn.compose import ColumnTransformer

print('ISNULL SUM PER COLUMN:')
print(df.isnull().sum())

df.drop_duplicates(inplace = True)
df.dropna(inplace = True)

print('\nOUTLIER INDEXES PER COLUMN:')
outliers_array = []
for name in df_numerical.columns.to_list():
    Q1 = df[name].quantile(0.05)
    Q3 = df[name].quantile(0.95)
    IQR = Q3 - Q1
    print(f'{name} IQR = {IQR}')
    logical_index_outliers = (df[name] < (Q1 - 1.5 * IQR)) |(df[name] > (Q3 + 1.5 * IQR))
    positives = np.where(logical_index_outliers == True)[0]
    print(f'{name} Outlier indexes: {positives}')

to_be_removed = ['Sales']

# log transform:
df['Advertising'] = np.log(df['Advertising'] + 1)

# ordinal encoding:
dictionary={"ShelveLoc": {"Bad":0, "Medium":1, "Good":2}}
df.replace(dictionary, inplace=True)

print('\nISNULL SUM ON SEL. COL.:', df['ShelveLoc'].isnull().sum())
```

```python
df.drop([df.index[]) # put the indexes outputted in the lists above
```

### Pipelines

```python
numeric_features = ['CompPrice', 'Income', 'Advertising',
                    'Population','Price','Age','Education']
# numeric transformer
numeric_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='median')),
    ('scaler', RobustScaler())])

# categorical transformer
categorical_features = ['Urban', 'US']
categorical_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='most_frequent')),
    ('categorical', OneHotEncoder())])

# selected transformer on selected column
preprocessor = ColumnTransformer(
    remainder = 'passthrough',
    transformers=[
        ('numeric', numeric_transformer, numeric_features),
        ('categorical', categorical_transformer, categorical_features),
        ('remove', 'drop', to_be_removed)])

X = preprocessor.fit_transform(df)
y = df['Sales']
# start training models.
```

Capping outlier rows with percentiles

```python
upper_lim = df['chol'].quantile(.95)
lower_lim = df['chol'].quantile(.05)

df.loc[(df['chol'] > upper_lim),'chol'] = upper_lim
df.loc[(df['chol'] < lower_lim),'chol'] = lower_lim
```

## Feature selection

Find importance of every feature

```python
# for csv or downloaded files
from sklearn.linear_model import LassoCV
import matplotlib.pyplot as plt
plt.figure(figsize=(12,5))

df_nosales = df.drop(['Sales'], axis=1)
df_nosales = pd.get_dummies(df_nosales)

lasso = LassoCV().fit(df_nosales, y)
importance = np.abs(lasso.coef_)
col_names = np.array(df_nosales.columns)
plt.bar(height=importance, x=col_names)
plt.title("Feature importance via coefficients")
plt.show()
```
